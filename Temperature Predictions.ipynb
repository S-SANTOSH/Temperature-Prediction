```python
model_max = ["new1","bra1","sou1","ukr1","fra1","swe1","jap1","pol1"]
country_max = [new_zealand2,brazil2,south_africa2,ukraine2,france2,sweden2,japan2,poloand2]
model_min = ["new2","bra2","sou2","ukr2","fra2","swe2","jap2","pol2"]
country_min = [new_zealand3,brazil3,south_africa3,ukraine3,france3,sweden3,japan3,poloand3]
def plot2(x,y,a , q,w,e):
    plt.figure(figsize=(25,15))
    gsl = gridspec.GridSpec(3,3)
    plt.subplots_adjust(hspace=None,wspace=0.4,bottom=None,top=None,right=None,left=None)
    axl = plt.subplot(gsl[i])
    plt.scatter(x,y,c="blue")
    plt.plot(x,a,c="red")
    plt.scatter(q,w,c="green")
    plt.plot(q,e,c="orange")
```

```python
def model_func(countri_max,countri_min,i):
    model_max[i] = LinearRegression()
    model_min[i] = LinearRegression()
    x_max=countri_max[:,1]
    y_max=countri_max[:,4]
    x_min=countri_min[:,1]
    y_min=countri_min[:,4]
    model_max[i].fit([x_max],[y_max])
    model_min[i].fit([x_min],[y_min])
    y_pred_min = model_min[i].predict([x_min])
    y_pred_max = model_max[i].predict([x_max])
    y_pred_max = y_pred_max[0]
    y_pred_min = y_pred_min[0]
    plot2(x_max,y_max,y_pred_max,x_min,y_min,y_pred_min)
    return(y_pred_max , y_pred_min)
```
